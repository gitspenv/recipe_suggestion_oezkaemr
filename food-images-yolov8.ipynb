```python
from ultralytics import YOLO
import os
import matplotlib.pyplot as plt
```

```python
data_path = '/kaggle/input/food-images/food_images/data.yaml'
```

```python
model = YOLO('yolov8l.pt')
```

```python
model.train(data=data_path, epochs=150, imgsz=640, batch=16, optimizer='AdamW', lr0=0.00008, momentum=0.9)
```

```python
test_images_path = '/kaggle/input/food-images/food_images/test/images' 

test_images = os.listdir(test_images_path)
print(f"Found {len(test_images)} test images.")

for img_name in test_images[:15]:
    img_path = os.path.join(test_images_path, img_name)
    results = model(img_path)
    
    result_img = results[0].plot()
    plt.figure(figsize=(10, 10))
    plt.imshow(result_img)
    plt.axis('off')
    plt.title(f"Predictions for {img_name}")
    plt.show()

```
